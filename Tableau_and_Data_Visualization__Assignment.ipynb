Q1. What is Tableau?

 - Tableau is a Business Intelligence and data visualization tool used to analyze data and create interactive dashboards and reports. It helps organizations make data-driven decisions by identifying trends, patterns, and insights through visual analysis.

Q2. Explain the role of Tableau components.

 - a) Data Pane
Displays all data fields from the data source and is used to drag fields into visualizations.

 - b) Worksheet
Used to create individual charts and visualizations.

 - c) Dashboard
Combines multiple worksheets into a single interactive view.

 - d) Story
Presents a sequence of worksheets or dashboards to tell a data-driven story.

Q3. Difference between Dimensions and Measures.

 - Dimensions are categorical fields used to classify data (e.g., Country, Product).
Measures are numerical fields used for calculations (e.g., Sales, Profit).

Q4. Filters, Parameters, and Sets.

 - Filters limit the data shown in a view.
Parameters allow users to input values dynamically.
Sets define a subset of data based on conditions.
